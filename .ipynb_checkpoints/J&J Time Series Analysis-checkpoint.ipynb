### Common Function
To plot TS, ACF and PCAF

```r
ts_acf_pacf_plot <- function(data, title='Time Series'){
    plot( data, main = title)
    acf( data, main="ACF")
    pacf( data, main="PACF")
}
```

### Log-Return Transformation
##### To transform and stablize the variance

```r
data <- diff(log(JohnsonJohnson))
# options(repr.plot.width=4, repr.plot.height=3)
ts_acf_pacf_plot(data, "Log-return (mean zero) of Johnson&Johnson earnings per share")
```

<b>Interpretation:</b><br>
From an ACF graph, we can see the auto-correlation at lag - 4, 8, 12, etc. due to quarter cycle. So, we are going to take a seasonal difference with lag=4

### Adding Seasonal Differencing with lag = 4

```r
data <- diff(diff(log(JohnsonJohnson)),4)
ts_acf_pacf_plot(data, "Time Series after Seasonal and Non-Seasonal Differencing")
```

### Ljung-Box Test 

```r
Box.test(data, lag = log(length(data)))
```

<b>Interpretation:</b><br>
p-value is value is vary less compared to our significance level(0.05), so we can reject the null hypothesis that there is no autocorrelation between previous lags. Now, autocorrelation exists, lets analyze ACF and PACF plots.

### Auto-Correlation Function (ACF)

```r
acf(data)
```

<b>Interpretation:</b><br>
* Lag for non-seasonal moving average could be 0 or 1 after that it is decaying and with-in the blue region.
* Autocorrelation at lag-4 is there, it's value is quite close to blue region and if could be a significant value.

<b>Order Estimation:</b><br>
* Non-seasonal, MA(p) <- 0 or 1
* Season, MA(P) <- 0 or 1

### Partial-AutoCorrelation Function (PACF)

```r
pacf(data)
```

<b>Interpretation:</b><br>
* Non-seasonal auto-regressive order could be 0 or 1
* Season autoregressive order is having partial autocorrelation at lag-4 so it's order can also be 0 or 1


### Order Estimation Using SARIMA Model

```r
d=1 #Non-Season Differencing
DD=1 #Seasonal Differencing
per=4 # Season

# Estimating order by fitting multiple combinations of p,d,q,P,D,Q using SARIMA modelling
for(p in 1:2){
  for(q in 1:2){
    for(i in 1:2){
      for(j in 1:2){
        if(p+d+q+i+DD+j<=10){
        
        # SARIMA Modelling
          model<-arima(x=log(jj), order = c((p-1),d,(q-1)), seasonal = list(order=c((i-1),DD,(j-1)), period=per))
        
        # Ljung-Box test to check auto-correlation in residuals
          pval<-Box.test(model$residuals, lag=log(length(model$residuals)))
        
        # Sum of squared errors of residuals
          sse<-sum(model$residuals^2)
        
        # Printing the output
          cat(p-1,d,q-1,i-1,DD,j-1,"Seasonality=",per, 'AIC=', model$aic, ' SSE=',sse,' p-VALUE=', pval$p.value,'\n')
        }
      }
    }
  }
}
```

<b>Interpretation</b><br>
Our objective is to find a combination with
* High p-value
* Lowest AIC
* Lowest SSE
* Comply to Parsimonius principle <br>

According to above conditions, we rounded our answer to <b>(0,1,1,0,1,0).</b><br>


### Residual Analysis

```r
options(repr.plot.width=14, repr.plot.height=10)
sarima(log(jj),0,1,1,1,1,0,4)
```

<b>Interpretation:</b><br>
1. p-value of estimates indicates that values are statistically significant.
2. Standardized Residual chart indicate somewhat consistent behaviour across time like a white noise. It seems stationary.
3. Normal Q-Q plot indicates residual follows normal distribution
4. ACF indicates no significant auto-correlation at any lag.
5. series of p-values at each lag tells there is nothing left as an autocorrelation in the residuals.
